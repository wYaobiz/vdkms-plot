```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('data_2.xlsx', sheet_name='dockerVSoutside')
```

```python
df
```

```python
df.drop([10, 11], axis=0, inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

```python
df.loc[:, 'Clients'] = df.loc[:, 'AWS Clients'].astype(int) * df.loc[:, 'C'].astype(int)
df
```

```python
# df.drop([3, 9, 13], axis=0, inplace=True)
# df.reset_index(inplace=True, drop=True)
# df
```

```python
df.loc[df['Instance Type'] == '4*t2.xlarge, indy node outside', 'Instance'] = "4 separate instances"
df.loc[df['Instance Type'] == '1*t2.xlarge, von-network', 'Instance'] = "Dockerized single instance"
df
```

```python
melted_df = pd.melt(df.loc[:, ['Clients', 'Instance', 'Success']], id_vars=['Clients', 'Instance'])
melted_df
```

```python
import matplotlib.pyplot as plt

sns.set_theme(rc={'figure.figsize':(15, 5)})
plot = sns.lmplot(data=melted_df, x='Clients', y='value', hue='Instance', legend=True, markers=["o", 's'], height=6, aspect=2)
plot.set_xlabels('Clients (Number of AWS Instances * Number of process per Instance)')
plot.set_ylabels('Successful Txns')
plot.set_titles('Comparison between Dockerized Single Instance and 4 Running Instances based on Number of Clients (Running Time = 20 minutes, N = 10 txns/sec)')
plot.legend.set_title('')
# plot.title.set_text('Comparison between Dockerized Single Instance and 4 Running Instances based on Number of Clients (Running Time = 20 minutes, N = 10 txns/sec)')
# plot.set(ylabel='Successful Txns', xlabel='Clients (Number of AWS Instances * Number of process per Instance)')
# _ = plot.legend(title='')


for x, y in zip(melted_df.loc[melted_df['Instance'] == '4 separate instances', 'Clients'], melted_df.loc[melted_df['Instance'] == '4 separate instances', 'value']):
    plt.text(x = x, # x-coordinate position of data label
    # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
    y = y + 2500, # y-coordinate position of data label, adjusted to be 150 below the data point
    s = '{:.0f}'.format(x), # data label, formatted to ignore decimals
    )

for x, y in zip(melted_df.loc[melted_df['Instance'] == 'Dockerized single instance', 'Clients'], melted_df.loc[melted_df['Instance'] == 'Dockerized single instance', 'value']):
    plt.text(x = x, # x-coordinate position of data label
    # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
    y = y - 8500, # y-coordinate position of data label, adjusted to be 150 below the data point
    s = '{:.0f}'.format(x), # data label, formatted to ignore decimals
    )
```
